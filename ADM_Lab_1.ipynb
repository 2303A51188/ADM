```python
xpoints = df['Age_Group']
ypoints = df['Heart_Attack_Incidence']
plt.plot(xpoints, ypoints)
plt.show()
```

Both Adults and Youth have equal chances of having heartattacks

Smoker Adults have high chance of having heartattack,
Non-Smoker Adults have 0 chance of having heartattack,
Most of youth are former smokers
Former smokers may or may not have heartattack



```python
sns.relplot(data=df, x='Age_Group',y='Heart_Attack_Incidence',hue='Physical_Activity_Level',kind='scatter')
```

Adults with low Physical Activity have high chance of having Heart attacks

Adults with Moderate Physical Activity have low chance of having Heart attacks


```python
sns.relplot(data=df, x='Age_Group',y='Stress_Level',hue='Physical_Activity_Level',style='Heart_Attack_Incidence')
```

Youth & Adults with Moderate Stress Level and Low Physical Activiy have high chance of having Heart Attacks
