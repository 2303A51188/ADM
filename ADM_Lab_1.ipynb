```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('heart_attack_germany.csv')
data.head()
```

```python
#checking missing values
print(data.isnull().sum())
```

no missing values found

```python
#min-max normalization

from sklearn.preprocessing import MinMaxScaler

#initializing MinMaxScaler
scaler = MinMaxScaler()

#selecting columns to normalize
columns_to_normalize = ['Year','BMI','Alcohol_Consumption','Family_History','Hypertension','Cholesterol_Level','Diabetes','Air_Pollution_Index','Region_Heart_Attack_Rate']

#Applying min-max normalization
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

data[columns_to_normalize].head()
```

```python
#Standardization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#Apply Standardization to numerical coumns
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

data[columns_to_normalize].head()


```
