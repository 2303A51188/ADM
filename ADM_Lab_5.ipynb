```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('titanic.csv')

data.head()
```

```python
data.columns
```

```python
#2. Data Overview
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
#3. Handling Missing Values

data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

#Verifying that missing values are handled
print(data.isnull().sum())
```

```python
# 4. Correlation Analysis

correlation_matrix = data.corr()

plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',fmt='.2f',cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#5. Visualizing Trends, Patterns, and Outliers

#Trend Analysis - scatter plot for Age vs Fare to analyze trends

plt.figure(figsize=(8,6))
sns.scatterplot(x='Age',y='Fare',data=data,hue='2urvived',palette='Set1')
plt.title('Age vs Fare: Trend Analysis')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python

# Visualizing Distribution of Numerical Features - Distribution plot for Age

plt.figure(figsize=(8,6))
sns.histplot(data['Age'],bins=30,kde=True,color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python

#Visualizing Outliers - Box plot to visualize outliers in Fare
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Fare'],color='lightcoral')
plt.title('Fare Outliers')
plt.xlabel('Fare')
plt.show()
```

```python

#Count plot for Pclass (Passenger class)
plt.figure(figsize=(8,6))
sns.countplot(x='Pclass',data=data,palette='Set2')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python

# Count plot for Survived (Survival distribution)

plt.figure(figsize=(8,6))
sns.countplot(x='2urvived',data=data,palette='Set2')
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
#
```
