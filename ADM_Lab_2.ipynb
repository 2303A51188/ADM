```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('apple_sales_2024.csv')
print(data.head())
```

```python
print(data.describe())
```

```python
data.keys()
```

```python
print(data.isnull().sum())
```

```python
data.dtypes
```

```python
print(data.describe(include=['object']))
```

```python
#histogram
plt.hist(data['iPhone Sales (in million units)'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('iPhone Sales Distribution')
plt.xlabel('Sales (in million units)')
plt.ylabel('Frequency')
plt.show()
```

```python
#histogram
plt.hist(data['Mac Sales (in million units)'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Mac Sales Distribution')
plt.xlabel('Sales (in million units)')
plt.ylabel('Frequency')
plt.show()
```

```python
#barchart
plt.figure(figsize=(10, 6))
data['iPhone Sales (in million units)'].plot(kind='bar',color='skyblue')
plt.title('iPhone Sales Distribution')
plt.xlabel('Sales (in million units)')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
#boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data['Wearables (in million units)'].dropna(),patch_artist=True,boxprops=dict(facecolor='skyblue'))
plt.title('Box plot of Wearables Sales Distribution')
plt.xlabel('Sales (in million units)')
plt.ylabel('Frequency')
plt.show()
```

```python
#pairplot
sns.pairplot(data[['Mac Sales (in million units)','Wearables (in million units)','iPhone Sales (in million units)']].dropna())
plt.show()
```

```python
#correlation map
correlation_matrix = data[['Mac Sales (in million units)','Wearables (in million units)','iPhone Sales (in million units)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#countplot
sns.countplot(data=data, x='iPhone Sales (in million units)',palette='Set2')
plt.title('iPhone Sales Distribution')
plt.xlabel('Sales (in million units)')
plt.ylabel('Frequency')
plt.show()
```

```python
#violinplot
sns.violinplot(data=data, y='iPhone Sales (in million units)',x='Region',palette='muted')
plt.title('iPhone Sales Distribution by Region')
plt.show()
```
